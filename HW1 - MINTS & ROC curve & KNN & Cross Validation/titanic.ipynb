```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import matplotlib
%matplotlib inline
```

```python
data = pd.read_csv('titanic/train.csv')
```

```python
Y = data["Survived"] 
X = data[["Pclass","Sex","Age","SibSp","Parch", "Fare"]]
```

```python
missing_values = X.isnull()
```

```python
print(sum(missing_values["Pclass"]))
print(sum(missing_values["Sex"]))
print(sum(missing_values["Age"]))
print(sum(missing_values["SibSp"]))
print(sum(missing_values["Parch"]))
print(sum(missing_values["Fare"]))
```

```python
ages = [i for i in X["Age"] if str(i) != "nan"]
average = sum(ages)/len(ages)
average
```

```python
fare = [i for i in X["Fare"] if str(i) != "nan"]
averageFare = sum(fare)/len(fare)
averageFare
```

```python
X["Age"] = X["Age"].fillna(average)
X["Sex"] = pd.get_dummies(X['Sex'], drop_first=True)
X["Age"] = X["Age"] / X["Age"].max()
X["Fare"] = X["Fare"] / X["Fare"].max()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=101)
```

```python
logistic = LogisticRegression() #0.77
logistic.fit(X_train, y_train)
```

```python
logistic.score(X_test, y_test)
```
