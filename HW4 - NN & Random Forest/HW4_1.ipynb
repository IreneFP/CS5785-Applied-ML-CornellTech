```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def getLosses(filename):
    with open(filename, "r") as dFile:
        lines = dFile.readlines()[:5000]
    return [float(d.split()[-1]) for d in lines]
```

```python
losses = getLosses("losses.csv")
```

```python
plt.plot(losses)
plt.title("Losses over Time")
plt.xlabel("Time")
plt.ylabel("Losses")
```

```python
changingRateLoss = getLosses("changingRateLosses.csv")
```

```python
plt.figure(figsize = [12,7])
plt.plot(changingRateLoss)
plt.title("Losses over Time with Different Learning Rates After Every 1000 Iterations")
plt.xlabel("Time")
plt.ylabel("Losses")
```

```python
plt.figure(figsize = [15,7])
plt.plot(losses)
plt.plot(changingRateLoss)
plt.legend(['Loss', 'changingRateLoss'])
plt.title("Original Loss vs. Loss While Decreasing Learning Rate")
plt.ylabel("Loss")
```

```python
files = ["layers/{} layers.csv".format(i) for i in range(2,10)]
files[-3] = "losses.csv"
lossesData = [getLosses(filename) for filename in files]
```

```python
plt.figure(figsize = [15,10])
layers = ["{} layers".format(i) for i in range(2,10)]
plt.title("Losses over Time for Network with Varying Hidden Layers")
plt.ylabel("Loss")
plt.xlabel("Time")

minForLayer = {}
for layer, data in enumerate(lossesData):
    plt.plot(data)
    minForLayer[layer + 2] = data[-1]

    
plt.legend(layers)
print(list(sorted(minForLayer.items(), key=lambda x: x[1])))
```
